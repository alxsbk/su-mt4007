### Apartment Prices Data

```python
import pandas as pd
df = pd.read_csv("Booli_sold.csv")
df.head()
```

### Simple Calulation of ppsqm for each apartment sold 

```python
# calculate ppsqm
df['ppsqm'] = df['soldPrice'] / df['livingArea']
df.head()
```

In this table, ppsqm is in the 29th column.

### Rank the top 5 most expensive appartments w.r.t ppsqm

```python
#sort data
top_5_expensive = df.sort_values(by='ppsqm', ascending=False).head(5)
top_5_expensive
```

In this table, the order is in descending order from top down.

### Avg Sqaure Price in Ekhagen

```python
# calculate the average ppsqm in Ekhagen
average_ppsqm_ekhagen = df['ppsqm'].mean()
average_ppsqm_ekhagen
```

The average square price in Ekhagen is: 58759.38

### Highlight an aspect of the data that you find interesting

Something I found interesting about the data in Ekhagen is that there are only four streets. Usually, the street an apartment is on, even within the same neighborhood, can affect the price. However, as we see here, the price per square meter (ppsqm) difference between the four streets isn’t significant. This could indicate that other factors play a bigger role in determining the ppsqm in Ekhagen, such as size, distance to the ocean, and so on.

```python
# function to extract street name without number
def extract_street_name(address):
    return address.split()[0] 

# apply the function to extract the street name
df['street_name'] = df['location.address.streetAddress'].apply(extract_street_name)

# group by street name and calculate the average price per square meter
average_ppsqm = df.groupby('street_name')['ppsqm'].mean().reset_index()

sorted_streets = average_ppsqm.sort_values(by='ppsqm', ascending=False)

print(sorted_streets)
```

## Swedish Election 2018

```python
# Load your data, specifying the separator as ';'
df = pd.read_csv("2018_R_per_kommun.csv", sep=';')
df.head()

```

### Calculate the total number of legitimate votes (Giltiga Röster) in Stockholm during the election

```python
# filter the data to only include rows where the county name is "Stockholms län"
stockholm_data = df[df['LÄNSNAMN'] == "Stockholms län"]

# calculate the total number of legitimate votes in Stockholm
total_legitimate_votes = stockholm_data['RÖSTER GILTIGA'].sum()

print("Total legitimate votes in Stockholm:", total_legitimate_votes)
```

### In which municipality did the social democratic party (Social demokraterna, S) garner the hightest voting percentage?

```python
# convert the 'S' column to numeric
df['S'] = df['S'].astype(str).str.replace(',', '.').astype(float)


# find the row with the highest percentage of 'S' votes
max_s_row = df.loc[df['S'].idxmax()]

highest_s_municipality = max_s_row['KOMMUNNAMN']
highest_s_percentage = max_s_row['S']

print(f"The Social Democratic Party received the highest voting percentage in {highest_s_municipality} with {highest_s_percentage:.2f}% of votes.")
```

### Rank the (3) municipialities with highest participation (valdeltagande). Present this in the form of a table.

```python
df['VALDELTAGANDE'] = df['VALDELTAGANDE'].astype(str).str.replace(',', '.').astype(float)
# sort the DataFrame by VALDELTAGANDE in descending order and select the top 3 rows
top_3_participation = df[['KOMMUNNAMN', 'VALDELTAGANDE']].sort_values(by='VALDELTAGANDE', ascending=False).head(3)


print(top_3_participation)
```

Here we can see that Lomma ís the highest with 93.86%, second Habo with 93.35%, and third Vellinge with 93.13%.

## Predicting Strokes 

In this part i chose to analyse the Age Distribution by Stroke Incidence.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv("stroke-data.csv")

# Filter data into two groups: individuals with and without stroke
stroke_yes = df[df['stroke'] == 1]['age']
stroke_no = df[df['stroke'] == 0]['age']

# Plot the data with updated parameter
plt.figure(figsize=(10, 6))
plt.boxplot([stroke_no, stroke_yes], tick_labels=['No Stroke', 'Stroke'])
plt.xlabel('Stroke Incidence')
plt.ylabel('Age')
plt.title('Age Distribution by Stroke Incidence')
plt.show()
```

The next step is to derive a simplier model with less parameters which are hopefully statiscally significant. We will use backward selection.
In backward elimination, you start with all predictors and remove them one at a time, keeping the ones that improve the model in terms of minimizing AIC. AIC helps find the model that balances being simple and explaining the data well, with a lower AIC indicating a better model.

```python
selected_features, final_model = my_code.backward_elimination(X, y)
print("Selected Features:", selected_features)
my_code.model_diagnostics(final_model, X, y)
```


After performing backward out the same result we se that we get a model with the parameters: 'Malicacid', 'Alcalinity_of_ash', 'Color_intensity', '0D280_0D315_of_diluted_wines', 'Proline. Let also check the Variance Inflation Factor (VIF) which  quantifies multicollinearity among explanatory variables in a regression model by measuring how much the variance of a regression coefficient is inflated due to collinearity with other variables. A VIF of 1 indicates no collinearity, while values greater than 5 suggest problematic multicollinearity, which can increase the standard errors of coefficients, making it harder to assess their individual contributions to the response.

```python
vif_df = my_code.calculate_vif_for_model(final_model, X)
vif_df
```

The Variance Inflation Factor values are all below 2 (except for const, which is expected to be high due to its role as an intercept), indicating low multicollinearity among the predictors. This suggests that the explanatory variables are relatively independent, and their inclusion in the model does not cause instability in the coefficient estimates. The results affirm the model's robustness and reliability in interpreting individual effects.

# Key Parameters

Malic Acid (Coef: 0.1176, p = 0.004)
Malic acid, associated with tartness in wine, has a significant positive relationship with alcohol content. For every 1-unit increase in malic acid, alcohol content rises by approximately 0.12 units. This effect is statistically significant at the 95% level, indicating it plays a meaningful role.

Alcalinity of Ash (Coef: -0.0326, p = 0.019)
Ash alcalinity, a measure of mineral content, shows a negative association with alcohol. A 1-unit increase in alcalinity corresponds to a 0.03-unit decrease in alcohol content. This effect is small but significant, suggesting that higher mineral content might slightly dilute alcohol levels.

Color Intensity (Coef: 0.1521, p < 0.001)
Color intensity, often linked to richness and bold flavors, is strongly and positively associated with alcohol content. A 1-unit increase in intensity is associated with a 0.15-unit rise in alcohol. This result is highly significant and emphasizes the importance of color in signaling alcohol levels.

OD280/OD315 of Diluted Wines (Coef: 0.1666, p = 0.026)
This measure relates to phenolic content (compounds affecting flavor and mouthfeel). Its positive coefficient suggests that higher phenolic content leads to a 0.17-unit increase in alcohol content. While its effect is smaller compared to color intensity, it is statistically significant and relevant.

Proline (Coef: 0.0011, p < 0.001)
Proline, an amino acid important for grape ripening, is positively associated with alcohol. Each unit increase in proline leads to a 0.0011-unit increase in alcohol. While the coefficient is small due to proline’s scale, it is highly significant, marking it as an important predictor.

# Conclusion
This analysis highlights color intensity, proline, and malic acid as the most impactful factors on alcohol content, with phenolic content and alcalinity of ash also contributing. The model strikes a balance between simplicity and explanatory power.
